## Penggunaan Subplots:

- Digunakan untuk membuat plot dalam beberapa axes. plt.subplots() sendiri memiliki dua output yaitu figure dan axes. Serta memiliki input berupa jumlah axes yang ingin dibuat (baris dan kolom).


```python
fig, ax =  plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'], df['Physical Sciences'], color='blue')
```

```python
plt.plot(df['Year'], df['Physical Sciences'], c='r', marker='o', ls='--')
```

Axes memiliki bentuk seperti matriks, maka untuk memilih axes tersebut dapat memanggil seperti matriks.
- ax[0] = untuk axes yg berada di titik [0], jika hanya terdapat 1 baris atau kolom
- ax[0,1] = untuk axes yg berada di baris 0 dan       row ke 1


```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                        figsize=(8,5))

ax[0].plot(df['Year'], df['Physical Sciences'], color='blue')
ax[1].plot(df['Year'], df['Computer Science'], color='red')
```

## Mengubah Warna

- Untuk merubah warna kita dapat memasukkannya di dalam fungsi plot() dengan input plot(color=’tipe_warna’).
- List-list warna yang dapat dipilih bisa dilihat di https://jonasjacek.github.io/colors/, selain input berupa warna, kita juga dapat memasukan dalam bentuk HEX

## Mengubah Jenis Garis

## Mengubah Marker Data Points
- Untuk merubah jenis marker kita dapat memasukkannya di dalam fungsi plot() dengan input plot(marker=’tipe_marker’).
- List-list jenis garis dan marker yang dapat dipilih bisa dilihat di https://matplotlib.org/2.1.1/api/_as_gen/matplotlib.pyplot.plot.html

## Mengubah Judul
- penggunaan title dalam satu axes dapat menggunakan fungsi plt.title(), salah satu input dalam fungsi tersebut yang dapat kita ubah adalah tipe dan ukuran font.


```python
fig, ax =  plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'], df['Physical Sciences'], color='blue')

plt.title('Physical Sciences Bachelors over year', fontsize=16)
```

set_title:
 - Mirip dengan penggunaan title dalam satu axes, perbedaannya terdapat pada definisi axes yang digunakan dan nama fungsinya.
 - Kita harus mengatur axes yg digunakan (misal ax[0]) baru kita gunakan fungsi ax[0].set_title()


```python
fig, ax =  plt.subplots(2, 1, figsize=(8,5))

ax[0].plot(df['Year'], df['Physical Sciences'], color='blue')
ax[1].plot(df['Year'], df['Computer Science'], color='red')

ax[0].set_title("Physical Sciences Bachelors over year", fontsize=16)
ax[1].set_title("Computer Sciences Bachelors over year", fontsize=16)

fig.tight_layout(pad=3.0)
```

## Mengubah label
- untuk mengubah label memiliki bentuk format yang sama dengan mengubah title, perbedaannya ada pada nama fungsi yang dipakai, yaitu plt.xlabel() untuk menamai label di X axis dan plt.ylabel() untuk menamai label di Y axis.

```python
fig, ax =  plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'], df['Physical Sciences'], color='blue')
plt.title('Physical Sciences Bachelors over year', fontsize=16)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage Bachelors', fontsize=12)
```

set_xlabel dan set_ylabel:
- untuk mengubah label memiliki bentuk format yang sama dengan mengubah title (set_title), perbedaannya ada pada nama fungsi yang dipakai, yaitu ax.set_xlabel() untuk menamai label di X axis dan ax.set_ylabel() untuk menamai label di Y axis.


```python
fig, ax =  plt.subplots(2, 1, figsize=(8,5))

ax[0].plot(df['Year'], df['Physical Sciences'], color='blue')
ax[1].plot(df['Year'], df['Computer Science'], color='red')

ax[0].set_title("Physical Sciences Bachelors over year", fontsize=16)
ax[1].set_title("Computer Sciences Bachelors over year", fontsize=16)

ax[1].set_xlabel('Year', fontsize=12)
ax[0].set_ylabel('Percentage Bachelors', fontsize=12)
ax[1].set_ylabel('Percentage Bachelors', fontsize=12)

fig.tight_layout(pad=3.0)
```

## Menggunakan X-Y limit
- xlim dan ylim digunakan untuk memberikan limit pada sumbu x dan y. Penggunaan sintaksnya dengan memasukkan input nilai limit awal dan nilai limit akhir.
- plt.xlim(limit awal, limit akhir)


```python
fig, ax =  plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'], df['Physical Sciences'], color='blue')
plt.title('Physical Sciences Bachelors over year', fontsize=16)

plt.xlim(1990,2010)
plt.ylim(0,50)
```

set_xlim dan set_ylim:
- xlim dan ylim digunakan untuk memberikan limit pada sumbu x dan y. Penggunaan sintaksnya dengan memasukkan input nilai limit awal dan nilai limit akhir.
- ax.set_xlim(limit awal, limit akhir)




```python
fig, ax =  plt.subplots(2, 1, figsize=(8,5),
                        sharex=True,
                        sharey=True)

ax[0].plot(df['Year'], df['Physical Sciences'], color='blue')
ax[1].plot(df['Year'], df['Computer Science'], color='red')

ax[0].set_xlim(1990, 2010)
ax[0].set_ylim(0, 50)

ax[1].set_xlim(1990, 2010)
ax[1].set_ylim(0, 50)

ax[0].set_xticks(range(1990, 2014, 2))
```

## Menggunakan legenda
- legenda digunakan untuk membedakan beberapa data yang ada di dalam suatu plot. penggunaan sintaksnya, dapat memasukkan input letak legenda yang diinginkan.
- plt.legend(loc=’upper left’)


```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color='blue', label='Physical Sciences')
plt.plot(df['Year'], df['Computer Science'], color='red', label='Computer Science')

plt.legend(loc=1)
```

- legenda digunakan untuk membedakan beberapa data yang ada di dalam suatu plot. penggunaan sintaksnya, dapat memasukkan input letak legenda yang diinginkan.
- plt.legend(loc=’upper left’)


```python
fig, axs =  plt.subplots(2, 1, figsize=(8,5))

axs[0].plot(df['Year'], df['Physical Sciences'], color='blue', label='Physical Sciences')
axs[0].plot(df['Year'], df['Computer Science'], color='red', label='Computer Science')
axs[0].legend(loc='lower right')

axs[1].plot(df['Year'], df['Physical Sciences'], color='blue')
axs[1].plot(df['Year'], df['Computer Science'], color='darkorange')
```

## Menambahkan teks

- teks dapat digunakan untuk menambahkan suatu keterangan pada plot yang kita buat, untuk memperjelas suatu data atau insight
- plt.text(x, y, s=’text’)


```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color='blue')
plt.plot(df['Year'], df['Computer Science'], color='red')

plt.text(1975, 25, s='computer science',
         fontsize=12,
         color='red',
         rotation=63)
```

- teks dapat digunakan untuk menambahkan suatu keterangan pada plot yang kita buat, untuk memperjelas suatu data atau insight
- ax.text(x, y, s=’text’)


```python
fig, axs =  plt.subplots(2, 1, figsize=(8,5))

axs[0].plot(df['Year'], df['Physical Sciences'], color='blue')
axs[1].plot(df['Year'], df['Computer Science'], color='red')

axs[1].text(1972,20, s='computer science',
            fontsize=12,
            color='red',
            rotation=45)
```

## Menambahkan anotasi

Anotasi biasa digunakan untuk menunjukan atau memberi keterangan pada hal penting, penggunaan sintaksnya dengan memasukkan beberapa input yaitu, letak titik x & y akan ditunjuk, letak titik x dan y dari text yang akan ditampilkan, teks yang akan ditulis, dan keterangan jenis panah yang akan digunakan


```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Computer Science'], color='red')
plt.ylim(10, 50)

plt.annotate('maximum',xy=(1983,37), xytext=(1983+5, 37+5), fontsize=16,
             arrowprops=dict(ec='black', # color arrow
                             arrowstyle='simple',))
```

## Menambahkan Horizontal Line
- Horizontal Line ini biasa digunakan untuk menunjukan suatu nilai batas dari nilai tertentu, penggunakan sintaksnya dengan memasukan nilai dari y axis yang akan kita tampilkan garisnya, ada beberapa pilihan yang dapat ditambahkan seperti warna, jenis garis, dan juga tebal garis

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color='red')

plt.annotate('threshold',xy=(1970,26))
plt.axhline(y=25, color='blue', ls='--', lw=1.5)
```

## Menambahkan Vertical Line
- Vertical Line ini biasa digunakan untuk menunjukan suatu nilai yang akan di highlight pada plot, penggunakan sintaksnya dengan memasukan nilai dari x axis yang akan kita tampilkan garisnya, ada beberapa pilihan yang dapat ditambahkan seperti warna, jenis garis, dan juga tebal garis


```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color='red')

plt.annotate('important event',xy=(1986,15))
plt.axvline(x=1985, color='blue', ls='--', lw=1.5)
```
